```python
import pandas as pd
import matplotlib.pyplot as plt
# load dataset
raisinSample = pd.read_csv(r'C:\Users\jjras\Documents\Raisin_Dataset.csv')
# Display the first 6 rows
firstSixRows = raisinSample.head(6)
print("First 6 rows of the DataFrame:")
print(firstSixRows)
```

```python
# Number of Rows and columns
print("\n The number of rows: ", len(raisinSample))
print("The number of columns: ", len(raisinSample.columns))
```

```python
# Find the dispersion in the dataset
raisinSampleNum = raisinSample.loc[:, raisinSample.columns != 'Class']
print("\nDispersion (Standard Deviation):\n", raisinSampleNum.std())
print('\nVariance: ', raisinSampleNum.var())
```

```python
# The numerical attributes
print("\nThe numerical attributes and summary in the dataset are: ")
attributes = raisinSample.head(0)
print(attributes)
```

```python
# A summary of the attributes
stats = raisinSample.describe()
print(stats)
```

```python
# The datatype of each attribute
print("\nThe datatypes of each attribute: ")
print(raisinSample.dtypes)
```

```python
# The labels of the data set
print("\nThe labels and labels datatype of the dataset are: ")
print(raisinSample.columns)
```

```python
# Not sure how many features or relevancy there are in
# this dataset but eliminated so columns that felt less important
# Dataset with only five features
raisinSample5 = raisinSample.loc[:, ['Area', 'MajorAxisLength', 'MinorAxisLength', 'ConvexArea', 'Perimeter']].copy()
# Dataset with only two features
raisinSample2 = raisinSample.loc[:, ['Area', 'Perimeter']].copy()
```
